```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def read_coefficients(file_path):
    with open(file_path, "r") as file:
        content = file.readlines()
        coefficients = [line.strip().split(",") for line in content]
        return [[float(coef) for coef in line] for line in coefficients]

def quadratic(time, coefficients):
    a, b, c = coefficients
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    coefficients_list = read_coefficients("/content/File2.txt")

    for coefficients in coefficients_list:
        time_values = np.linspace(0, 10, 50)
        temp_values = quadratic(time_values, coefficients)

        sns.set_style("darkgrid")
        plt.plot(time_values, temp_values, label=f'Temperature (coefficients: {coefficients})')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.title('Quadratic Temperature over Time')
    plt.legend(["set1[34,56,1]","set2[3.4,5.6,7]","set3[0.1,4.0,3]"])
    plt.show()

if __name__ == "__main__":
    main()
```

```python
def quadratic(time):
  a=0.5
  b=0.3
  c=1

  tempreture=a*(time**2)+b*(time)+c
  return tempreture

def quadratic2(time):
  a=34
  b=67
  c=1

  tempreture=a*(time**2)+b*(time)+c
  return tempreture

def quadratic3(time):
  a=100
  b=200
  c=23

  tempreture=a*(time**2)+b*(time)+c
  return tempreture

def main():
  time_values=np.linspace(0,10,100)
  time_values2=np.linspace(0,20,100)
  time_values3=np.linspace(0,30,100)

  temp_val1=quadratic(time_values)
  temp_val2=quadratic(time_values2)
  temp_val3=quadratic(time_values3)
  sns.set_style("darkgrid")
  plt.plot(time_values, temp_val1, label='Temperature',color="red")
  plt.plot(time_values, temp_val2, label='Temperature',color="royalblue")
  plt.plot(time_values, temp_val3, label='Temperature',color="green")
  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.title('Quadratic Temperature over Time')
  plt.legend(["set1[0.5,0.3,1]","set2[34,67,1]","set3[100,200,23]"])
  plt.show()


if __name__=="__main__":
  main()
```
